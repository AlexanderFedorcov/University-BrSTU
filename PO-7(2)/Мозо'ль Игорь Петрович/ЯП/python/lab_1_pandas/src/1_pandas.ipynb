```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import matplotlib
import seaborn as sns
```

```python
data = pd.read_csv('dataset.csv', delimiter='\t')
data
```

```python
data[data.isnull().any(1)]
```

```python
data = data.drop(['Z_CostContact', 'Z_Revenue'], axis = 1)
data
```

```python
print('После: ', data.columns, sep='\n')
data = data.rename({'Year_Birth' : 'Age'}, axis=1)
print('До: ', data.columns)
```

```python
fig, axs =plt.subplots(ncols=2, figsize=(10, 4))

MaritalResponseData = data[['Marital_Status', 'Response']]
MaritalData = data['Marital_Status']

response = MaritalResponseData.groupby('Marital_Status')['Response'].sum() / MaritalData.size * 100
values = MaritalData.value_counts() / MaritalData.size * 100

df = pd.DataFrame([response, values]).transpose()
df.plot.bar(stacked=True, ax=axs[0])

pie = plt.pie(values, labels=values.index, explode=[0.05] * len(values), autopct="%.2f%%")
```

```python
fig, axs = plt.subplots(ncols=1, figsize=(10, 4))
age = (2021 - data['Age'])
sns.histplot(x=age, kde=True)
```

```python
fig, axs = plt.subplots(ncols=4, figsize=(15, 4))

kidhome = data['Kidhome']
kidhome = kidhome.value_counts() / kidhome.size * 100
axs[0].bar(kidhome.index, kidhome.values)

teenhome = data['Teenhome']
teenhome = teenhome.value_counts() / teenhome.size * 100
axs[1].bar(teenhome.index, teenhome.values)

response = data['Response']
response = response.value_counts() / response.size * 100
axs[2].bar(response.index, response.values)

sns.kdeplot(data['Income'], shade=True)
```

```python
fig, axs = plt.subplots(ncols=4, figsize=(17, 4))

MaritalData = data[['Marital_Status', 'Income']]
for MD in MaritalData['Marital_Status'].unique():
    status = MaritalData[MaritalData['Marital_Status'] == MD][:100]
    sns.kdeplot(status['Income'], shade=True, warn_singular = False, ax=axs[0])
    
EducationData = data[['Education', 'Income']]
for ED in EducationData['Education'].unique():
    status = EducationData[EducationData['Education'] == ED][:100]
    sns.kdeplot(status['Income'], shade=True, warn_singular = False, ax=axs[1])
    
KidhomeData = data[['Kidhome', 'Income']]
for KD in KidhomeData['Kidhome'].unique():
    status = KidhomeData[KidhomeData['Kidhome'] == KD][:100]
    sns.kdeplot(status['Income'], shade=True, warn_singular = False, ax=axs[2])
    
ResponseData = data[['Response', 'Income']]
for i in range(2):
    response = ResponseData[ResponseData['Response'] == i][:100]
    sns.kdeplot(response['Income'], shade=True, warn_singular = False, ax=axs[3])
```
